# 企業への投資、就職に役立つ計算
  注目企業数社のIR情報を入力する
## columnとその計算式
増収率 = ( 売上高 - 前年度売上高 ) / 前年度売上高

当期純益 = 売上高 - ( 給与 + 税金 + その他 )

増益率 = ( 当期純益 - 前年度当期純益 ) / 前年度当期純益

総資産 = 資本 + 負債

ROA = 当期純益 / 総資産

売上高利益率 = 当期純益 / 売上高

価格維持 =

           if ( 某社の増収率 ) >= 0 then 某社の増収率 / 増収率の数社間最大値
           if ( 某社の増収率 ) < 0 then - ( 某社の増収率 ) / 増収率の数社間最小値

部門維持 =

           if ( 某社の増益率 ) >= 0 then 某社の増益率 / 増益率の数社間最大値
           if ( 某社の増益率 ) < 0 then - ( 某社の増益率 ) / 増益率の数社間最小値

元手活用 =

           if ( 某社のROA ) >= 0 then 某社のROA / ROAの数社間最大値
           if ( 某社のROA ) < 0 then - ( 某社のROA ) / ROAの数社間最小値

会社成長 =
           
           if ( 某社の売上高利益率 ) >= 0 then 某社の売上高利益率 / 売上高利益率の数社間最大値
           if ( 某社の売上高利益率 ) < 0 then - ( 某社の売上高利益率 ) / 売上高利益率の数社間最小値

A = ( 価格維持 + 部門維持 + 元手活用 + 会社成長 ) / 4

投資環境 = 4 * ( 某社のA - Aの数社間最小値 ) / ( Aの数社間最大値 - Aの数社間最小値 )

活用還元 = 2 * ( 1 - | 元手活用 | ) - 1

成長還元 = 2 * ( 1 - | 会社成長 | ) - 1

B = ( 価格維持 + 部門維持 + 活用還元 + 成長還元 ) / 4

労働環境 = 4 * ( 某社のB - Bの数社間最小値 ) / ( Bの数社間最大値 - Bの数社間最小値 )

総合評価 = ( 投資環境 + 労働環境 ) / 2

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 入力情報 : 売上高(億円), 前年度売上高(億円), 当期純益(億円), 前年度当期純益(億円), 総資産(億円)
idx_input = ["企業名", "売上高(億円)", "前年度売上高(億円)", "当期純益(億円)", "前年度当期純益(億円)", "総資産(億円)"]

# 注目企業数社のIR情報を入力する
company_1 = ["日立製作所", 108811.150, 102646.02, 7038.70, 6708.64, 125014.14]
company_2 = ["ソニーグループ", 115398.37, 99215.13, 9436.22, 8884.06, 320412.22]
company_3 = ["三菱電機", 50036.94, 44767.58, 2139.08, 2034.82, 55825.19]
company_4 = ["キヤノン", 40314.14, 35133.57, 2439.61, 2147.18, 50955.30]
company_5 = ["NTT Data", 34901.82, 25519.06, 1499.62, 1429.79, 61581.94]
company_6 = ["オープンハウスグループ", 11484.84, 9526.86, 920.50, 778.84, 11986.68]
# 入力終わり

df_input = pd.DataFrame(data = [company_1, company_2, company_3, company_4, company_5, company_6], columns = idx_input) # data = [] の中を要変更
display(df_input)
```

```python
idx_output = ["企業名", "売上高(億円)", "前年度売上高(億円)", "増収率(%)", "当期純益(億円)", "前年度当期純益(億円)", "増益率(%)",
       "総資産(億円)", "ROA(%)", "売上高利益率(%)", "価格維持", "部門維持", "元手活用", "会社成長", "投資環境", "労働環境", "総合評価"]

df_output = pd.DataFrame(data=[], columns=idx_output)
df_output["企業名"] = df_input["企業名"]
df_output["売上高(億円)"] = df_input["売上高(億円)"]
df_output["前年度売上高(億円)"] = df_input["前年度売上高(億円)"]
df_output["増収率(%)"] = (df_input["売上高(億円)"] - df_input["前年度売上高(億円)"]) / df_input["前年度売上高(億円)"] * 100
df_output["当期純益(億円)"] = df_input["当期純益(億円)"]
df_output["前年度当期純益(億円)"] = df_input["前年度当期純益(億円)"]
df_output["増益率(%)"] = (df_input["当期純益(億円)"] - df_input["前年度当期純益(億円)"]) / df_input["前年度当期純益(億円)"] * 100
df_output["総資産(億円)"] = df_input["総資産(億円)"]
df_output["ROA(%)"] = df_input["当期純益(億円)"] / df_input["総資産(億円)"] * 100
df_output["売上高利益率(%)"] = df_input["当期純益(億円)"] / df_input["売上高(億円)"] * 100
df_output.loc[df_output["増収率(%)"] >= 0, "価格維持"] = df_output["増収率(%)"] / df_output["増収率(%)"].max()
df_output.loc[df_output["増収率(%)"] < 0, "価格維持"] = - df_output["増収率(%)"] / df_output["増収率(%)"].min()
df_output.loc[df_output["増益率(%)"] >= 0, "部門維持"] = df_output["増益率(%)"] / df_output["増益率(%)"].max()
df_output.loc[df_output["増益率(%)"] < 0, "部門維持"] = - df_output["増益率(%)"] / df_output["増益率(%)"].min()
df_output.loc[df_output["ROA(%)"] >= 0, "元手活用"] = df_output["ROA(%)"] / df_output["ROA(%)"].max()
df_output.loc[df_output["ROA(%)"] < 0, "元手活用"] = - df_output["ROA(%)"] / df_output["ROA(%)"].min()
df_output.loc[df_output["売上高利益率(%)"] >= 0, "会社成長"] = df_output["売上高利益率(%)"] / df_output["売上高利益率(%)"].max()
df_output.loc[df_output["売上高利益率(%)"] < 0, "会社成長"] = - df_output["売上高利益率(%)"] / df_output["売上高利益率(%)"].min()
A = (df_output["価格維持"] + df_output["部門維持"] + df_output["元手活用"] + df_output["会社成長"]) / 4
df_output["投資環境"] = 4 * (A - A.min()) / (A.max() - A.min())
B = (df_output["価格維持"] + df_output["部門維持"] + 2 * (1 - df_output["元手活用"].abs()) - 1 + 2 * (1 - df_output["会社成長"].abs()) - 1) / 4
df_output["労働環境"] = 4 * (B - B.min()) / (B.max() - B.min())
df_output["総合評価"] = (df_output["投資環境"] + df_output["労働環境"]) / 2

display(df_output)
```
